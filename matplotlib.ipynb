```python
df=pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2016'],width=0.8)
```

```python

import numpy as np
plt.bar(np.arange(df.shape[0]) - 0.3,df['2015'],width=0.25)
plt.bar(np.arange(df.shape[0]) ,df['2016'],width=0.25)
plt.bar(np.arange(df.shape[0]) + 0.3,df['2017'],width=0.25)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.title('season wise comparison of best players')
plt.xlabel('Players')
plt.ylabel('Runs scored')
plt.savefig('comparison.png')
```

```python
import numpy as np
plt.bar(np.arange(df.shape[0]) - 0.3,df['2015'],width=0.25)
plt.bar(np.arange(df.shape[0]) ,df['2016'],width=0.25)
plt.bar(np.arange(df.shape[0]) + 0.3,df['2017'],width=0.25)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.title('season wise comparison of best players')
plt.xlabel('Players')
plt.ylabel('Runs scored')
plt.xticks(rotation='vertical')
```

**stacked bar graph**

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016'] + df['2017'],label='2016')

plt.show()
```
